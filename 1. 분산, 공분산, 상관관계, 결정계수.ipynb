#### 분산(deviation)

* 데이터가 일정 기준값(일반적으로 평균값)을 기준으로, 얼마나 퍼져있는지 알려주는 값

#### 공분산(covariance)
* X, Y 두 가지 데이터가 있을 때, 이 데이터들이 어떤 관계를 가지는지 알려주는 값.  
  양수 : X가 증가할 떄, Y도 증가하는 경향을 보임  
  음수 : X가 증가할 떄, Y는 감소하는 경향을 보임  
  0 : 서로 상관없음

#### pearson 상관계수(r이라고 보통 표기)
* 앞서 말한, 분산과 공분산의 개념을 이용하여 새롭게 구한 상관계수

$$
\begin{align}
r = \frac{Covarince}{\sqrt{X deviation * Y deviation}}
\end{align}
$$

#### 결정계수 (coefficient of determination, R) - 앞선 pearson 상관계수를 제곱한 수
$$
R = r^2
$$
